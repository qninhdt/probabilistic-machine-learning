```python
N = 1000
p = np.array([0.1, 0.3, 0.6])
values = np.array([100, 200, 300])
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x=np.random.choice(values, N, p=p), histnorm="probability"))
```

```python
# central limit theorem visualisation
M = 1000
K = 100
fig = go.Figure()
x = [np.mean(np.random.choice(values, K, p=p)) for _ in range(M)]
fig.add_trace(
    go.Histogram(
        x=x,
        histnorm="probability",
    )
)

from scipy.stats import norm

mean_x = values @ p
std_x = np.sqrt((values - mean_x) ** 2 @ p) / np.sqrt(K)
left_x = mean_x - 3 * std_x
right_x = mean_x + 3 * std_x
print(mean_x, std_x)
# draw normal distribution line
fig.add_trace(
    go.Scatter(
        x=np.linspace(left_x, right_x, 100),
        y=norm.pdf(np.linspace(left_x, right_x, 100), mean_x, std_x),
        mode="lines",
        name="normal distribution",
    )
)
fig.show()
```
