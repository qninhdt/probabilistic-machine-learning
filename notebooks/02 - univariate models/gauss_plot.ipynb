```python
from scipy.stats import norm

gaussian_pdf = norm(0, 1).pdf
gaussian_cdf = norm(0, 1).cdf

x = np.linspace(-4, 4, 1000)
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=x, y=gaussian_pdf(x), mode="lines", name="pdf", line=dict(color=COLOR_LIST[0])
    )
)

fig.add_trace(
    go.Scatter(
        x=x, y=gaussian_cdf(x), mode="lines", name="cdf", line=dict(color=COLOR_LIST[1])
    )
)

fig.update_layout(
    xaxis=dict(title="x"),
    yaxis=dict(title="y"),
    title="Gaussian PDF and CDF",
    showlegend=True,
)

alpha = 0.1

right_x = norm(0, 1).ppf(1 - alpha / 2)
left_x = norm(0, 1).ppf(alpha / 2)

left_x = x[x <= left_x]
fig.add_trace(
    go.Scatter(
        x=left_x,
        y=gaussian_pdf(left_x),
        mode="lines",
        name="pdf",
        line=dict(color=COLOR_LIST[0]),
        fill="tozeroy",
    )
)

right_x = x[x >= right_x]
fig.add_trace(
    go.Scatter(
        x=right_x,
        y=gaussian_pdf(right_x),
        mode="lines",
        name="pdf",
        line=dict(color=COLOR_LIST[0]),
        fill="tozeroy",
    )
)
```

```python
def mixed_gaussian_pdf(x):
    return 0.5 * norm.pdf(x, 0, 0.5) + 0.5 * norm.pdf(x, 2, 0.5)


def mixed_gaussian_cdf(x):
    return 0.5 * norm.cdf(x, 0, 0.5) + 0.5 * norm.cdf(x, 2, 0.5)


fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=x,
        y=mixed_gaussian_pdf(x),
        mode="lines",
        name="pdf",
        line=dict(color=COLOR_LIST[0]),
    )
)

fig.add_trace(
    go.Scatter(
        x=x,
        y=mixed_gaussian_cdf(x),
        mode="lines",
        name="cdf",
        line=dict(color=COLOR_LIST[1]),
    )
)
```
