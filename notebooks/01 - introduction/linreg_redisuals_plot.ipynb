```python
N = 100
```

```python
x = np.linspace(0, 1, N + 1)
x = x[:, np.newaxis]
```

```python
ones = np.ones((N + 1, 1))
X = np.hstack((ones, x))
```

```python
w = np.array([-2, 1 / 7.0])
y = w[0] * x + w[1] * np.square(x)
y = y + np.random.normal(0, 0.2, (N + 1, 1))
```

```python
X.shape, y.shape
```

```python
w = np.linalg.lstsq(X, y, rcond=None)[0]
w
```

```python
estimated_y = X @ w
```

```python
plt.plot(X[:, 1], y, "o")
plt.plot(X[:, 1], estimated_y, "-")
plt.show()

for x0, y0, y_hat in zip(X[:, 1], y, estimated_y):
    plt.plot([x0, x0], [y0, y_hat], "k-")
plt.plot(X[:, 1], y, "o")
plt.plot(X[:, 1], estimated_y, "-")
plt.show()
```
